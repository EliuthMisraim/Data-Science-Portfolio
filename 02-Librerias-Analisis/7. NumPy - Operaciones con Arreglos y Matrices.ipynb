## 1. Arreglos, Métodos y Operaciones

Un arreglo de NumPy (`ndarray`) es una cuadrícula de valores, todos del **mismo tipo** (usualmente números). Al forzar que todos los datos sean del mismo tipo, NumPy optimiza el uso de la memoria de la computadora, logrando cálculos matemáticos entre 10 y 50 veces más rápidos que las listas tradicionales de Python.

NumPy incluye cientos de **operaciones y métodos** preconstruidos para álgebra lineal, estadística y trigonometría (ej. `np.mean()`, `np.sum()`, `np.max()`).

```python
import numpy as np

# 💡 Ejemplo: Listas vs NumPy Arrays
lista_python = [1, 2, 3, 4, 5]
arreglo_numpy = np.array([1, 2, 3, 4, 5])

print(f"Tipo del arreglo: {type(arreglo_numpy)}")

# Operaciones súper rápidas: Sumar 10 a todos los elementos
# En una lista normal, esto daría error. En NumPy, es matemática pura:
arreglo_sumado = arreglo_numpy + 10

print(f"Arreglo original: {arreglo_numpy}")
print(f"Arreglo sumado +10: {arreglo_sumado}")
```

## 2. Broadcasting e Indexación



* **Broadcasting:** Es la magia oculta de NumPy. Describe cómo la librería trata arreglos de diferentes formas durante las operaciones aritméticas. Por ejemplo, si sumas un número solo (escalar) a un arreglo completo, NumPy "transmite" (broadcasts) ese número a todos los elementos del arreglo automáticamente.
* **Indexación y Asignación:** Funciona de manera similar a las listas de Python (usando corchetes `[inicio:fin:paso]`), pero a un nivel multidimensional.

```python
# 💡 Ejemplo: Indexación en acción
arreglo_base = np.array([10, 20, 30, 40, 50])

# Extraer el último elemento usando indexación negativa
ultimo = arreglo_base[-1]

# Extraer una "rebanada" (slicing) de los tres primeros
primeros_tres = arreglo_base[0:3]

print(f"Último elemento: {ultimo}")
print(f"Primeros tres: {primeros_tres}")
```

## 📝 Ejercicios Prácticos

A continuación, resolveremos tres retos algorítmicos utilizando las ventajas de NumPy.

### Ejercicio 1: Invertir un Arreglo
**Objetivo:** Escribir una función `invierte_array` que reciba un arreglo NumPy y lo devuelva invertido (el primer elemento pasa a ser el último) usando indexación.

```python
# --- Solución Ejercicio 1 ---
def invierte_array(array):
    # Utilizamos slicing [inicio:fin:paso]. Un paso de -1 recorre hacia atrás.
    return array[::-1]

# Ejemplo de uso
array_original = np.array([1, 2, 3, 4, 5])
array_invertido = invierte_array(array_original)

print("--- Ejercicio 1 ---")
print(f"Original:  {array_original}")
print(f"Invertido: {array_invertido}")
```

### Ejercicio 2: Conversor de Temperatura (Vectorización)
**Objetivo:** Escribir una función `fahrenheit_a_centigrados` que convierta un arreglo entero de grados Fahrenheit a Centígrados usando la fórmula $C = (5(F - 32)) / 9$. Redondear a dos decimales con `np.round()`.

```python
# --- Solución Ejercicio 2 ---
def fahrenheit_a_centigrados(array):
    # La matemática se aplica a todo el arreglo simultáneamente (Vectorización)
    centigrados = np.round((5 * (array - 32)) / 9, 2)
    return centigrados

array_temperaturas = np.array([0, 12, 45.21, 34, 99.91, 32])

print("--- Ejercicio 2 ---")
print(f"Valores Fahrenheit: {array_temperaturas}")
print(f"Valores Centígrados: {fahrenheit_a_centigrados(array_temperaturas)}")
```

### Ejercicio 3: Unión de Arreglos (Teoría de Conjuntos)
**Objetivo:** Escribir una función `union_de_arreglos` que reciba dos arreglos y devuelva uno nuevo, ordenado y sin elementos repetidos, utilizando `np.union1d`.

```python
# --- Solución Ejercicio 3 ---
def union_de_arreglos(array1, array2):
    # union1d compara ambos arreglos, une sus elementos y elimina duplicados
    union = np.union1d(array1, array2) # <-- Corregido: un solo np.
    return union

# Arreglos con al menos 3 elementos y 1 en común (el 10 y el 20)
array_A = np.array([0, 10, 20])
array_B = np.array([20, 10, 30])

print("--- Ejercicio 3 ---")
print(f"Arreglo 1: {array_A}")
print(f"Arreglo 2: {array_B}")
print(f"Unión:     {union_de_arreglos(array_A, array_B)}")
```

## 🔍 Explicación de los Algoritmos

Estos tres ejercicios demuestran por qué NumPy es la base de la Ciencia de Datos:

1. **Indexación Inteligente (Slicing):** En el Ejercicio 1, no usamos bucles `for` para reordenar los números. La sintaxis `[::-1]` le dice a la memoria de la computadora que simplemente lea las direcciones al revés. Costo computacional: casi cero.
2. **Vectorización pura:** En el Ejercicio 2, restamos 32 y multiplicamos por 5. Si hiciéramos esto con una lista nativa de Python, daría un *TypeError*. NumPy aplica la fórmula matemática a cada elemento del arreglo al mismo tiempo.
3. **Matemática de Conjuntos:** En el Ejercicio 3, usamos `np.union1d()`. NumPy tiene operaciones de conjuntos (unión, intersección, diferencia) súper optimizadas en lenguaje C por debajo del capó, lo que es vital para la limpieza de bases de datos y cruces de información.

## 🚀 Nivel Pro: Type Hinting con NumPy

Escribir código profesional en Python implica decirle a otros desarrolladores qué tipo de datos esperan nuestras funciones. Sin embargo, decir que un dato es una simple "lista" no es suficiente en Data Science. 

Para documentar funciones que usan NumPy de forma profesional, utilizamos la librería `numpy.typing` (o indicamos `np.ndarray` en nuestro código). Veamos cómo se agruparían tus tres ejercicios en una sola "Librería de Utilidades" lista para producción:

```python
import numpy as np

class NumpyUtilsPro:
    """
    Clase de utilidades matemáticas implementando operaciones vectorizadas con NumPy.
    Demuestra el uso de Type Hinting nativo de la librería (np.ndarray).
    """

    @staticmethod
    def invertir_arreglo(array: np.ndarray) -> np.ndarray:
        """Invierte el orden de un arreglo NumPy usando Slicing optimizado."""
        return array[::-1]

    @staticmethod
    def fahrenheit_a_celsius(array_f: np.ndarray) -> np.ndarray:
        """Convierte un arreglo de Fahrenheit a Celsius vectorizadamente."""
        return np.round((5 * (array_f - 32)) / 9, 2)

    @staticmethod
    def unir_arreglos(arr1: np.ndarray, arr2: np.ndarray) -> np.ndarray:
        """Retorna la unión matemática (sin duplicados y ordenada) de dos arreglos."""
        return np.union1d(arr1, arr2)

# --- Automatización de Pruebas ---
if __name__ == "__main__":
    print("=== 🧪 Pruebas Unitarias Pro ===")
    utils = NumpyUtilsPro()
    
    a1 = np.array([100, 200, 300])
    a2 = np.array([300, 400, 500])
    temp_f = np.array([32, 212]) # Congelación y Ebullición
    
    print(f"1. Inversión: {utils.invertir_arreglo(a1)}")
    print(f"2. Conversión 32F y 212F a Celsius: {utils.fahrenheit_a_celsius(temp_f)}")
    print(f"3. Unión Conjuntos: {utils.unir_arreglos(a1, a2)}")
```

## 🔍 Explicación de la Versión Profesional

1. **Tipado Estricto (`np.ndarray`):** Al especificar que los parámetros son de tipo `np.ndarray`, los editores de código (como VSCode) activarán el autocompletado exclusivo de NumPy y nos advertirán si intentamos pasar una lista normal de Python por accidente.
2. **Métodos Estáticos (`@staticmethod`):** Como estas funciones matemáticas no necesitan guardar un "estado" interno ni recordar variables de un objeto, usamos el decorador `@staticmethod`. Esto permite usarlas directamente como herramientas sin necesidad de instanciar la clase con `utils = NumpyUtilsPro()` si estuviéramos importándolo desde otro archivo.

## 📚 Glosario de NumPy



Para dominar la terminología de este módulo, aquí tienes los conceptos clave del ecosistema numérico de Python:

* **Broadcasting:** Mecanismo potente de NumPy que permite realizar operaciones aritméticas entre arreglos de diferentes formas o tamaños, "transmitiendo" el arreglo más pequeño a través del más grande para que tengan formas compatibles.
* **Escalar (Scalar):** En el contexto de NumPy, se refiere a un único número (una matriz de dimensión cero), a diferencia de un arreglo que contiene múltiples números.
* **Indexación (Indexing) / Slicing:** Técnica para acceder a elementos individuales o subconjuntos (rebanadas) de un arreglo. En NumPy, la indexación es multidimensional, permitiendo filtrar filas y columnas usando la sintaxis `[inicio:fin:paso]`.
* **Ndarray (N-dimensional array):** La estructura de datos central de NumPy. Es una colección de elementos (generalmente números) del mismo tipo, indexados por una tupla de enteros no negativos. Puede tener 1 dimensión (vector), 2 dimensiones (matriz) o N dimensiones (tensor).
* **Vectorización (Vectorization):** La práctica de expresar operaciones matemáticas aplicables a arreglos enteros sin escribir bucles `for` en Python. El cálculo se delega a código precompilado y altamente optimizado en C, resultando en ejecuciones casi instantáneas.
* **Unión de Conjuntos (`union1d`):** Operación que compara dos arreglos unidimensionales y devuelve un nuevo arreglo ordenado que contiene los elementos únicos presentes en cualquiera de los dos arreglos originales.
