## 1. Tablas de Verdad y lgebra Booleana
El 谩lgebra booleana es la base de la l贸gica computacional. Utiliza valores de Verdadero (`True`) y Falso (`False`). Los operadores l贸gicos nos permiten combinar estas condiciones:

* **`and` (Y):** Devuelve `True` solo si **ambas** condiciones son verdaderas.
* **`or` (O):** Devuelve `True` si **al menos una** condici贸n es verdadera.
* **`not` (NO):** Invierte el valor (de `True` a `False` y viceversa).

```python
# Ejemplos de lgebra Booleana
condicion_1 = True
condicion_2 = False

print(f"True AND False: {condicion_1 and condicion_2}")  # Falso, porque una es falsa
print(f"True OR False: {condicion_1 or condicion_2}")    # Verdadero, porque hay al menos una verdadera
print(f"NOT True: {not condicion_1}")                    # Falso, invierte el valor
```

## 2. Operadores Condicionales y la instrucci贸n `IF`
Los operadores condicionales (o relacionales) comparan dos valores y devuelven un valor booleano (`True` o `False`). Los m谩s comunes son: `>`, `<`, `>=`, `<=`, `==` (igual a), y `!=` (diferente de).

La instrucci贸n `if` ejecuta un bloque de c贸digo **煤nicamente** si la condici贸n evaluada es verdadera.

```python
edad = 18

# La instrucci贸n IF
if edad >= 18:
    print("Acceso permitido: Eres mayor de edad.")
```

## 3. La instrucci贸n `IF`, `ELSE` y `ELIF`
Para programas m谩s complejos, necesitamos evaluar m煤ltiples escenarios exclusivos:

* **`if`:** La primera condici贸n a evaluar.
* **`elif` (Else If):** Se eval煤a si el `if` anterior fue falso. Puedes poner tantos como necesites.
* **`else`:** El camino por defecto. Se ejecuta si **ninguna** de las condiciones anteriores fue verdadera. No lleva condici贸n.

```python
calificacion = 85

if calificacion >= 90:
    print("Excelente trabajo. Calificaci贸n: A")
elif calificacion >= 80:
    print("Buen trabajo. Calificaci贸n: B")
elif calificacion >= 70:
    print("Aprobado. Calificaci贸n: C")
else:
    print("Necesitas mejorar. Reprobado.")
```

```python
entrada_usuario = "veinte" # Simulamos que el usuario escribi贸 texto en lugar de un n煤mero

try:
    numero = float(entrada_usuario)
    print(f"El n煤mero ingresado es {numero}")
except ValueError:
    print("Error: Por favor, ingresa un valor num茅rico v谩lido, no texto.")
```

##  Ejercicio Pr谩ctico: Calculadora de N贸mina con Horas Extras

**Consigna:**
Desarrolle un programa en Python que calcule el pago semanal de un trabajador bas谩ndose en las horas laboradas y su tarifa por hora, incorporando un sistema de pago de horas extras y validaci贸n de datos.

**Requerimientos del c贸digo:**
1.  Solicitar al usuario las horas laboradas y la tarifa mediante `input()`.
2.  El pago normal aplica para las primeras 40 horas.
3.  Si trabaja entre 41 y 50 horas, el excedente sobre 40 se paga con un **10% de bono** (1.1x).
4.  Si trabaja m谩s de 50 horas, las primeras 10 extras se pagan al 1.1x, y el excedente sobre 50 se paga con un **20% de bono** (1.2x).

**Casos de prueba:**
> * Prueba 1: `40` horas, `$10/h` -> Resultado: `400.0 USD`
> * Prueba 2: `45` horas, `$10/h` -> Resultado: `455.0 USD` (400 + 55)

> * Prueba 4: `-5` horas o tarifa `texto` -> Resultado: `Par谩metros incorrectos`

```python
pago_total = 0
cadena_horas_laboradas = input("Ingrese el n煤mero de horas laboradas en una semana: ")
cadena_tarifa_por_hora = input("Ingrese su tarifa de pago por hora: ")

# L贸gica principal del ejercicio:
try:
    horas_trabajadas = float(cadena_horas_laboradas)
    tarifa_hora = float(cadena_tarifa_por_hora)
    
    # 1. Validaci贸n de l贸gicas negativas (lgebra Booleana)
    if horas_trabajadas < 0 or tarifa_hora < 0:
        print("Par谩metros incorrectos")
    else:
        # 2. Control de flujo para el c谩lculo (Condicionales)
        if horas_trabajadas <= 40:
            pago_total = horas_trabajadas * tarifa_hora
            
        elif horas_trabajadas <= 50:
            pago_total = (40 * tarifa_hora) + ((horas_trabajadas - 40) * tarifa_hora * 1.1)
            
        else:
            pago_total = (40 * tarifa_hora) + (10 * tarifa_hora * 1.1) + ((horas_trabajadas - 50) * tarifa_hora * 1.2)
        
        # 3. Formateo y salida
        pago_total = round(pago_total, 1)
        print(f"El monto total a pagar es: {pago_total} USD")

# 4. Manejo de excepciones
except ValueError:
    print("Par谩metros incorrectos")
```

##  Explicaci贸n del Algoritmo

El programa combina todo lo aprendido en el m贸dulo en una arquitectura de **validaci贸n y c谩lculo en cascada**:

3.  **C谩lculo Escalonado (`if / elif / else`)**:

    * **Tramo 2 (`elif`)**: Si llega a este punto, significa que las horas son estrictamente *mayores* a 40, pero se eval煤a si son `<= 50`. Calcula el sueldo base (40 horas) y le suma el excedente pagado al `1.1`.
    * **Tramo 3 (`else`)**: Si llega aqu铆, supera las 50 horas. Asegura el pago de las primeras 40 horas base, las 10 horas intermedias al `1.1`, y todas las restantes al `1.2`.

##  Versi贸n Profesional: Funciones y Cl谩usulas de Guarda (Guard Clauses)

En un entorno de trabajo real, el c贸digo no se escribe "suelto" en el archivo. Se encapsula en **funciones** para poder reutilizarlo, probarlo f谩cilmente y mantenerlo organizado. Adem谩s, se aplican principios de "Clean Code" (c贸digo limpio).

### Ventajas de esta versi贸n:
* **Modularidad:** Separamos la l贸gica matem谩tica de la interacci贸n con el usuario (los `input` y `print`).
* **Cl谩usulas de guarda (Guard Clauses):** Evaluamos los errores primero y "salimos" r谩pido de la funci贸n (con `return`). Esto evita anidar demasiados `if/else` y hace que el c贸digo sea m谩s plano y f谩cil de leer.
* **Tipado est谩tico (Type Hinting):** Ayuda a otros programadores (y a tu editor de c贸digo) a saber qu茅 tipo de datos espera y devuelve la funci贸n.

```python
def calcular_nomina(horas_str: str, tarifa_str: str) -> str:
    """
    Calcula el pago semanal aplicando bonos por horas extras.
    Retorna un string con el resultado formateado o un mensaje de error.
    """
    # 1. Cl谩usula de guarda: Intentar la conversi贸n de tipos
    try:
        horas = float(horas_str)
        tarifa = float(tarifa_str)
    except ValueError:
        return "Par谩metros incorrectos"
    
    # 2. Cl谩usula de guarda: Validar l贸gicas de negocio (sin n煤meros negativos)
    if horas < 0 or tarifa < 0:
        return "Par谩metros incorrectos"
    
    # 3. L贸gica principal (ya con la certeza de que los datos son limpios)
    if horas <= 40:
        pago_total = horas * tarifa
    elif horas <= 50:
        pago_total = (40 * tarifa) + ((horas - 40) * tarifa * 1.1)
    else:
        pago_total = (40 * tarifa) + (10 * tarifa * 1.1) + ((horas - 50) * tarifa * 1.2)
        
    return f"El monto total a pagar es: {round(pago_total, 1)} USD"

# --- Interacci贸n con el usuario (Punto de entrada del programa) ---
if __name__ == "__main__":
    print("===  Calculadora de N贸mina Pro ===")
    input_horas = input("Ingrese el n煤mero de horas laboradas en una semana: ")
    input_tarifa = input("Ingrese su tarifa de pago por hora: ")
    
    # Llamamos a la funci贸n e imprimimos su resultado directamente
    print(calcular_nomina(input_horas, input_tarifa))
```

2.  **Uso de `return` en lugar de `print`**: Una funci贸n matem谩tica rara vez deber铆a imprimir cosas en pantalla. Su trabajo es calcular y *devolver* (`return`) el resultado. As铆, el resultado puede ser guardado en una base de datos, enviado por API o, en este caso, impreso despu茅s.
3.  **Salida temprana (Early Return)**: Observa c贸mo tan pronto como detectamos un error (en el `except` o en el `if` de n煤meros negativos), usamos `return "Par谩metros incorrectos"`. Cuando Python lee un `return`, sale inmediatamente de la funci贸n. Esto nos ahorra tener que poner todo el c贸digo de c谩lculo dentro de un bloque `else` gigante.
