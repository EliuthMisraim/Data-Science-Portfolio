## 1. Conoce la instrucci√≥n `while`

El ciclo `while` (que significa "mientras" en ingl√©s) repite un bloque de c√≥digo **mientras una condici√≥n espec√≠fica sea verdadera (`True`)**. 

Es ideal cuando no sabemos exactamente cu√°ntas veces necesitamos repetir una acci√≥n, pero s√≠ sabemos bajo qu√© condici√≥n debe detenerse. 

**‚ö†Ô∏è La regla de oro del `while`:** Siempre debes asegurarte de que la condici√≥n en alg√∫n momento se vuelva falsa (por ejemplo, restando o sumando a un contador). De lo contrario, crear√°s un "bucle infinito" que congelar√° tu programa.

```python
# üí° Ejemplo b√°sico de While: Una cuenta regresiva

contador = 10
print("Iniciando cuenta regresiva...")

# El ciclo se ejecutar√° MIENTRAS el contador sea mayor a 0
while contador > 0:
    print(contador)
    contador -= 1  # Esto es crucial: reducimos el contador en 1 cada vez para evitar un bucle infinito

print("¬°Despegue! üöÄ")
```

## 2. Practica la instrucci√≥n `while`

Ahora que entendemos lo b√°sico, apliquemos el `while` a un problema matem√°tico donde no sabemos cu√°ntas veces tendremos que iterar hasta encontrar la respuesta.

**Objetivo:**
Desarrolle un programa en Python que mediante el uso de un ciclo `while` determine si un n√∫mero `n` (proporcionado por el usuario) puede ser expresado como $2^k$ (donde $k$ es entero).

**Reglas:**
* El valor de `n` debe ser un entero entre 2 y 100000.
* El resultado debe ser un booleano almacenado en la variable `es_potencia_de_2`.

**Casos de prueba:**
> * Si `n = 8` -> $2^3 = 8$ -> `True`
> * Si `n = 40` -> No hay $k$ entero -> `False`

```python
# 1. Captura y validaci√≥n estricta de datos
try:
    # Nota: Usamos 8 por defecto para no bloquear el notebook con un input manual, 
    # pero puedes cambiarlo por int(input("Ingrese n√∫mero: "))
    n = 8 
    print(f"N√∫mero ingresado a evaluar: {n}")
    
    if n < 2 or n > 100000:
        raise ValueError("¬°Error en el valor ingresado! ¬°Debe ser un n√∫mero entero entre 2 y 100000!")
except ValueError as e:
    raise ValueError("¬°Error en el valor ingresado! ¬°Debe ser un n√∫mero entero entre 2 y 100000!") from None

es_potencia_de_2 = False

# 2. L√≥gica del ciclo While
k = 0 
potencia = 1  # Inicialmente 2^0 

# Iteramos mientras la potencia calculada sea menor o igual al n√∫mero objetivo
while potencia <= n: 
    if potencia == n: 
        es_potencia_de_2 = True 
        break  # Rompemos el ciclo inmediatamente, ¬°ya encontramos la respuesta!
    
    k += 1 
    potencia = 2 ** k 

# 3. Impresi√≥n del resultado
print(f"Variable booleana es_potencia_de_2: {es_potencia_de_2}")
print(f"El n√∫mero {n} {'puede' if es_potencia_de_2 else 'no puede'} ser expresado como 2^k.")
```

## 3. Conoce la instrucci√≥n `for`

A diferencia del `while`, el ciclo `for` (que significa "para cada" en este contexto) se utiliza cuando **sabemos de antemano cu√°ntas veces queremos iterar**, o cuando queremos recorrer todos los elementos de una colecci√≥n (como una lista, un texto o un rango de n√∫meros). 

En Python, es muy com√∫n combinar `for` con la funci√≥n `range(inicio, fin)`, la cual genera una secuencia de n√∫meros autom√°ticamente.

```python
# üí° Ejemplo b√°sico 1: Iterar sobre los elementos de una lista
frutas = ["Manzana", "Banana", "Cereza"]
print("Lista de compras:")
for fruta in frutas:
    print(f"- {fruta}")

print("-" * 20)

# üí° Ejemplo b√°sico 2: Usar range() para repetir una acci√≥n un n√∫mero exacto de veces
# range(1, 4) genera los n√∫meros 1, 2 y 3 (el l√≠mite superior no se incluye)
print("Contando hasta 3:")
for numero in range(1, 4):
    print(numero)
```

## 4. Practica la instrucci√≥n `for`

Aplicaremos el ciclo `for` para crear una funci√≥n que dependa de un rango exacto de n√∫meros.

**Objetivo:**
Desarrolle un programa que calcule el **factorial** del n√∫mero que indique el usuario ($n!$). 

**Reglas:**
* El factorial de $n$ ($n!$) es la multiplicaci√≥n de $n$ por sus antecesores mayores o iguales a 1 (Ej. $4! = 4 \times 3 \times 2 \times 1 = 24$).
* Si el valor es inv√°lido (negativo, texto o fraccionario), debe mostrar "Error de captura".
* El resultado debe guardarse en la variable `factorial`.

```python
# Simulamos la entrada del usuario (puedes cambiar "4" por input())
cadena = "4"
factorial = 1  # Declarar la variable global

def calcular_factorial():
    global factorial  # Indicar que se usar√° la variable global
    try:
        # Verificar si la entrada es un n√∫mero entero no negativo
        if not cadena.isdigit():
            raise ValueError("Error de captura: La entrada no es un entero positivo.")
        
        n = int(cadena)
        if n < 0:
            raise ValueError("Error de captura: El n√∫mero debe ser positivo.")
        
        factorial = 1
        
        # Ciclo FOR: Iteramos desde 1 hasta 'n' (se suma 1 al final porque range excluye el l√≠mite)
        for i in range(1, n + 1):
            factorial *= i  # Acumulamos la multiplicaci√≥n: 1*1, luego 1*2, luego 2*3, luego 6*4...
            
        print(f"El factorial de {n} es {factorial}")
        
    except ValueError as e:
        print(f"Entrada no v√°lida: {e}")

# Llamada a la funci√≥n
calcular_factorial()
```

## üîç Explicaci√≥n de los Algoritmos

Ambos ejercicios demuestran c√≥mo elegir la herramienta correcta para el problema adecuado:

1.  **Algoritmo `while` (Potencia de 2):** * Se utiliza un `while` porque **no sabemos** qu√© tan grande ser√° `n` ni cu√°ntas veces tendremos que multiplicar el 2. 
    * Utilizamos la instrucci√≥n `break` para "romper" el ciclo inmediatamente al encontrar coincidencia, ahorrando recursos del sistema.
2.  **Algoritmo `for` (Factorial):** * Se utiliza un `for` junto con `range()`. **S√≠ sabemos** exactamente cu√°ntas veces iterar: $n$ veces.
    * La variable `factorial` act√∫a como un **acumulador**, actualizando su valor en cada paso.

## üöÄ Nivel Pro: Refactorizando el c√≥digo (Evitando `global`)

En el ejercicio del factorial utilizamos `global factorial`. En programaci√≥n profesional, **el uso de variables globales se evita** porque hace que el c√≥digo sea impredecible en programas grandes (si muchas funciones modifican la misma variable, es dif√≠cil rastrear errores).

Una funci√≥n debe ser como una "caja negra": recibe entradas, trabaja internamente y devuelve una salida (`return`), sin alterar el exterior. As√≠ se ver√≠a tu c√≥digo con "Clean Code":

```python
def calcular_factorial_pro(entrada: str) -> int:
    """Calcula el factorial de un n√∫mero retornando el valor directamente."""
    
    if not entrada.isdigit():
        raise ValueError("Error: La entrada no es un entero o es un decimal/negativo.")
    
    n = int(entrada)
    resultado_factorial = 1 # Variable local, vive y muere dentro de esta funci√≥n
    
    for i in range(1, n + 1):
        resultado_factorial *= i
        
    return resultado_factorial # Devolvemos el dato en lugar de imprimirlo

# --- Pruebas unitarias de nuestra funci√≥n ---
if __name__ == "__main__":
    casos_de_prueba = ["4", "5", "6", "7.5"] 
    
    for prueba in casos_de_prueba:
        try:
            resultado = calcular_factorial_pro(prueba)
            print(f"‚úÖ El factorial de {prueba} es: {resultado}")
        except ValueError as error:
            print(f"‚ùå Para el valor '{prueba}' -> {error}")
```

## üîç Explicaci√≥n de la Versi√≥n Profesional

Esta versi√≥n optimizada aplica principios de **Clean Code** (C√≥digo Limpio) para crear una funci√≥n robusta, segura y f√°cil de testear:

1. **Adi√≥s a la instrucci√≥n `global`**: El uso de variables globales es peligroso porque cualquier parte del programa podr√≠a modificarlas por accidente, causando errores dif√≠ciles de rastrear (conocidos como *efectos secundarios* o *side effects*). Al quitar `global`, protegemos nuestro entorno.
2. **Variables Locales**: La variable `resultado_factorial` nace cuando la funci√≥n es llamada y se destruye cuando la funci√≥n termina. Su existencia est√° encapsulada de forma segura dentro de la funci√≥n.
3. **Retorno de valores (`return`) en lugar de `print`**: Una funci√≥n matem√°tica no deber√≠a imprimir resultados directamente en pantalla. Su trabajo es calcular y *devolver* (`return`) el valor. Esto nos da la libertad de usar ese n√∫mero m√°s adelante (por ejemplo, guardarlo en un archivo excel, sumarlo a otra variable o, en este caso, imprimirlo con un formato personalizado).
4. **Automatizaci√≥n de Pruebas**: En lugar de ejecutar el programa 4 veces manualmente para probar los casos del ejercicio (`4`, `5`, `6` y `7.5`), creamos una lista `casos_de_prueba` y usamos un ciclo `for` para automatizar la evaluaci√≥n de todos los escenarios en un solo paso.
